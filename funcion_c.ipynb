```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def Function_C(T, H, T_min=0, alpha=1, beta=1):
    return H**(beta)/(T-T_min)**(alpha)
```

```python
T_min, T_max = 0, 50
temp = np.linspace(1, T_max, 100)
hum = np.linspace(0, 100, 100)

T, H = np.meshgrid(temp, hum)
Z = Function_C(T=T, H=H, T_min=T_min)
```

```python
fig = plt.figure()
ax = plt.axes(projection='3d')
ax.contour3D(T, H, Z, 1000, cmap='viridis')
ax.set_xlabel('Temperature')
ax.set_ylabel('Humity')
ax.set_zlabel('Constant C')
```

```python
ax.view_init(10,60)
fig
```

```python
v_t = np.array([20,25,30,35,40])
v_h = np.array(range(10,100,10))
```

```python
cte = np.zeros(shape=(len(v_t), len(v_h)), dtype='float64')
row = 0
col = 0
for t in v_t:
    for h in v_h:
        cte[row, col] = Function_C(T=t, H=h)
        col+=1
    row+=1
    col=0
```
